```python
import sys
from os import path
import pandas as pd
import pickle
from nltk import corpus
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from wordcloud import WordCloud, STOPWORDS
from sklearn.feature_extraction.text import CountVectorizer 
from sklearn.decomposition import NMF
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
with open('emilyspoems.pkl', 'rb') as picklefile:
    emilyspoems = pickle.load(picklefile)
```

```python
vectorizer = CountVectorizer(stop_words='english', ngram_range=(1,2), min_df=6, max_df=.1)
vectorizer.fit(emilyspoems)
x = vectorizer.transform(emilyspoems)
```

```python
x.shape
```

```python
x = pd.DataFrame(x.toarray(), index=emilyspoems, columns=vectorizer.get_feature_names())
```

```python
x.isna().sum().sum() / (x.shape[0] * x.shape[1])
```

```python
nmf_model = NMF(5, max_iter=1000, solver='mu', init='nndsvda')
doc_topic = nmf_model.fit_transform(x)
```

```python
topic_word = pd.DataFrame(nmf_model.components_.round(3),
             index = ["component_1",
                      "component_2",
                      "component_3",
                      "component_4",
                      "component_5",],
#                       "component_6"],
#                       "component_7",
#                       "component_8",
#                       "component_9",
#                      "component_10"],
             columns = vectorizer.get_feature_names())
topic_word
```

```python
def display_topics(model, feature_names, no_top_words, topic_names=None):
    for ix, topic in enumerate(model.components_):
        if not topic_names or not topic_names[ix]:
            print("\nTopic ", ix)
        else:
            print("\nTopic: '",topic_names[ix],"'")
        print(", ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))
```

```python
display_topics(nmf_model, vectorizer.get_feature_names(), 5)
```

```python
ex_label = [e[:30]+"..." for e in emilyspoems]
df_dominant_topic = pd.DataFrame(doc_topic.round(5),
             index = ex_label,
             columns = ["component_1",
                      "component_2",
                      "component_3",
                      "component_4",
                      "component_5",])
#                       "component_6"
#                       "component_7",])
#                       "component_8",
#                       "component_9",
#                      "component_10"])
df_dominant_topic
```

```python
J = df_dominant_topic.values
df_dominant_topic['Dominant_Topic'] = J.argmax(axis=1)
```

```python
import sys
from os import path
import numpy as np
from PIL import Image
from wordcloud import WordCloud, STOPWORDS

def random_color_func(word=None, font_size=None, position=None, orientation=None, font_path=None, random_state=None):
    h = int(360.0 * 165.0 / 255.0)
    s = int(100.0 * 255.0 / 255.0)
    l = int(100.0 * float(random_state.randint(60, 120)) / 255.0)

    return "hsl({}, {}%, {}%)".format(h, s, l)

def create_wordcloud(text):
     
    stopwords = set(STOPWORDS)
 
    wc = WordCloud(background_color="white",
                    max_words=21, 
                    prefer_horizontal=1.0,
                    stopwords=stopwords,
                    min_font_size=10,
                    width=3000,
                    height=1500,
                    color_func=random_color_func,
                    collocations=False,)
     
    wc.generate(text)
    plt.figure(figsize = (14, 8), facecolor = None) 
    plt.imshow(wc) 
    plt.axis("off") 
    plt.tight_layout() 
    plt.savefig("Word-Cloud-AllDocs.svg");
    
create_wordcloud(' '.join(emilyspoems))
```

```python
import seaborn as sns
import matplotlib.colors as mcolors
cols = [color for name, color in mcolors.TABLEAU_COLORS.items()]  # more colors: 'mcolors.XKCD_COLORS'

fig, axes = plt.subplots(5,1,figsize=(8,16), dpi=160, sharey=True)

for i, ax in enumerate(axes.flatten()):    
    df_dominant_topic_sub = df_dominant_topic.loc[df_dominant_topic.Dominant_Topic == i, :]
    doc_lens = df_dominant_topic_sub['Word_Count']
    ax.hist(doc_lens, bins = 100, color=cols[i])
    ax.tick_params(axis='y', labelcolor=cols[i], color=cols[i])
    sns.kdeplot(doc_lens, color="black", shade=False, ax=ax.twinx())
    ax.set(xlim=(0, 300), xlabel='Document Word Count')
    ax.set_ylabel('Number of Documents', color=cols[i])
    ax.set_title('Topic: '+str(i), fontdict=dict(size=16, color=cols[i]))

fig.tight_layout()
fig.subplots_adjust(top=0.90)
plt.xticks(np.linspace(0,300,5))
fig.suptitle('Distribution of Document Word Counts by Dominant Topic', fontsize=22)
plt.savefig("Word_Count_Dom_Topic.svg");
```
